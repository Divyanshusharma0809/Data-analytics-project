```python
#Filtering the Data
filter_Country_Region = abhi['Deaths']>10000
abhi[filter_Country_Region]
```

```python
# Drop unwanted columns
abhi = abhi.drop(['New cases', 'New recovered', 'New deaths'], axis=1)

# Display the first few rows to confirm
abhi.sample(5)
```

```python
px.bar(abhi.head(15), x = 'Country/Region',
       y = 'Deaths', height = 500)
```

```python
abhi.plot(x="Country/Region", y="Deaths", figsize=(18, 8))
plt.xlabel("Country/Region")
plt.ylabel("Deaths")
plt.title("Country/Region vs Deaths")
plt.show()
```

```python
abhi.plot(x="Country/Region", y="Deaths",color = "black",figsize=(28,9), marker = "o", kind = "line")
plt.xlabel("Country/Region")
plt.ylabel("Deaths")
plt.title("Country/Region vs Deaths")
plt.show()
```

```python
abhi.plot(x="Country/Region", y="Deaths",color = "black",figsize=(28,7), kind = "bar")
plt.xlabel("Country/Region")
plt.ylabel("Deaths")
plt.title("Country/Region vs Deaths")
plt.show()
```
